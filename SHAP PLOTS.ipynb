```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from skopt import BayesSearchCV
from sklearn.utils import resample
import shap

# Load the data
df = pd.read_excel("/content/dataset_and_validation.xlsx")
X = df[['pH', 'Dosage(mg)', 'Time(min)', 'PPM', 'Temp(K)', 'precursor']]
y = df['% Removal']

# Split the data
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.05, random_state=123)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1, random_state=123)

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)

# Hyperparameter optimization
param_space = {
    'C': (1, 10),
    'gamma': ['scale', 'auto'],
    'epsilon': (0.0001, 0.001),
    'kernel': ['poly', 'rbf']
}
opt = BayesSearchCV(
    SVR(),
    param_space,
    n_iter=100,
    cv=13,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=2,
    random_state=42
)
opt.fit(X_train_scaled, y_train)

print("Best parameters found: ", opt.best_params_)
print("Lowest RMSE found: ", np.sqrt(-opt.best_score_))

# Evaluate the best model
best_svr = opt.best_estimator_
y_pred = best_svr.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print("Test RMSE: ", np.sqrt(mse))

# Performance metrics
train_pred = best_svr.predict(X_train_scaled)
test_pred = best_svr.predict(X_test_scaled)
val_pred = best_svr.predict(X_val_scaled)

train_mae = mean_absolute_error(y_train, train_pred)
test_mae = mean_absolute_error(y_test, test_pred)
val_mae = mean_absolute_error(y_val, val_pred)

train_r2 = r2_score(y_train, train_pred)
test_r2 = r2_score(y_test, test_pred)
val_r2 = r2_score(y_val, val_pred)

print("Train Mean Absolute Error:", train_mae)
print("Test Mean Absolute Error:", test_mae)
print("Val Mean Absolute Error:", val_mae)

print("Train R-squared:", train_r2)
print("Test R-squared:", test_r2)
print("Val R-squared:", val_r2)

# Plot the training model
plt.scatter(y_train, train_pred, color='blue', label='Training Data')
plt.scatter(y_test, test_pred, color='red', label='Test Data')
plt.scatter(y_val, val_pred, color='green', label='Validation Data')
plt.plot(y_train, y_train, linestyle='--', color='black')
plt.xlabel('Actual % Removal')
plt.ylabel('Predicted % Removal')
plt.title('Actual vs Predicted % Removal (Training Data)')
plt.legend()
plt.show()
```

```python


# Bootstrapping
n_bootstrap_samples = 100
bootstrap_samples = []

for i in range(n_bootstrap_samples):
    X_resampled, y_resampled = resample(X_train_scaled, y_train, n_samples=len(X_train_scaled))
    bootstrap_samples.append((X_resampled, y_resampled))

# Calculate SHAP values for the original model using KernelExplainer
explainer_original = shap.KernelExplainer(best_svr.predict, X_train_scaled)
shap_values_original = explainer_original.shap_values(X_train_scaled)

# Create SHAP summary plot for the original model
shap.summary_plot(shap_values_original, X_train_scaled, feature_names=X.columns, title="Original Model SHAP Summary Plot")

# Calculate SHAP values for each bootstrapped model
shap_values_bootstrapped = []
for i, (X_resampled, _) in enumerate(bootstrap_samples):
    shap_values_bootstrapped.append(explainer_original.shap_values(X_resampled))
    # Optionally, you can create SHAP plots for each bootstrapped model here
```

```python
# Bootstrapping
import xlsxwriter
n_bootstrap_samples = 10
bootstrap_samples = []

# Create an Excel writer object
excel_writer = pd.ExcelWriter('/content/random_samples_with_predictions (1).xlsx', engine='xlsxwriter')

for i in range(n_bootstrap_samples):
    X_resampled, y_resampled = resample(X_train_scaled, y_train, n_samples=len(X_train_scaled))
    bootstrap_samples.append((X_resampled, y_resampled))

    # Calculate SHAP values for the bootstrapped sample
    explainer = shap.KernelExplainer(best_svr.predict, X_resampled)
    shap_values = explainer.shap_values(X_resampled)

    # Save SHAP values to Excel
    shap_values_df = pd.DataFrame(shap_values, columns=X.columns)
    shap_values_df.to_excel(excel_writer, sheet_name=f'SHAP_values_sample_{i+1}', index=False)

    # Create SHAP summary plot for the bootstrapped sample
    shap.summary_plot(shap_values, X_resampled, feature_names=X.columns, show=False)
    plt.title(f'SHAP Summary Plot for Bootstrapped Sample {i+1}')
    plt.savefig(f'/content/shap_plot_sample_{i+1}.png')
    plt.close()

    # Load the image file into the Excel sheet
    worksheet = excel_writer.sheets[f'SHAP_values_sample_{i+1}']
    worksheet.insert_image('G2', f'/content/shap_plot_sample_{i+1}.png')

# Close the Excel writer object and save the file
excel_writer.close()

print("SHAP values and plots for bootstrapped samples have been saved to Excel.")
```

```python
# Bootstrapping
n_bootstrap_samples = 100
bootstrap_samples = []
all_shap_values = []

# Create an Excel writer object
excel_writer = pd.ExcelWriter('/content/bootstrap_shap_values.xlsx', engine='xlsxwriter')

for i in range(n_bootstrap_samples):
    X_resampled, y_resampled = resample(X_train_scaled, y_train, n_samples=len(X_train_scaled))
    bootstrap_samples.append((X_resampled, y_resampled))

    # Calculate SHAP values for the bootstrapped sample
    explainer = shap.KernelExplainer(best_svr.predict, X_resampled)
    shap_values = explainer.shap_values(X_resampled)

    # Collect all SHAP values
    all_shap_values.append(shap_values)

    # Save SHAP values to Excel
    shap_values_df = pd.DataFrame(shap_values, columns=X.columns)
    shap_values_df.to_excel(excel_writer, sheet_name=f'SHAP_values_sample_{i+1}', index=False)

# Close the Excel writer object and save the file
excel_writer.close()

# Aggregate SHAP values
all_shap_values = np.array(all_shap_values)
mean_shap_values = np.mean(all_shap_values, axis=0)

# Create a combined SHAP summary plot
shap.summary_plot(mean_shap_values, X_train_scaled, feature_names=X.columns)
plt.title('Combined SHAP Summary Plot')
plt.show()

print("SHAP values and combined plot have been saved.")
```
