```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Loading the dataset from the excel
df = pd.read_excel("/content/Copy of Data Sec B(1).xlsx")
```

```python
# Separating the dependent variable (y)-target and independent variable(x)-features
x = df[['pH', 'Dosage(mg)', 'Time(min)', 'PPM', 'Temp(K)', 'precursor']]
y = df['% Removal']
```

```python
# Splitting the data into training and testing for test size 75%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.75, random_state=42)
```

```python

# Creating XGBoost model and training it with training data
xgb_model = XGBRegressor()
xgb_model.fit(x_train, y_train)
```

```python

# Predicting and finding the error of the model for 75% test size
train_pred = xgb_model.predict(x_train)
test_pred = xgb_model.predict(x_test)

tr_rmse = mean_squared_error(y_train, train_pred, squared=False)
te_rmse = mean_squared_error(y_test, test_pred, squared=False)
mae = mean_absolute_error(y_test, test_pred)
```

```python
print("Train root mean square error: ", tr_rmse)
print("Test root mean square error: ", te_rmse)
print("Mean absolute error for the y: ", mae)
```

```python
# Plotting the actual and predicted % removal for both testing and training data for 75% test size
plt.scatter(y_train, train_pred, color='red')
plt.plot(y_train, y_train, linestyle='-')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Training Data 25%)')
plt.scatter(y_test, test_pred, color='blue')
plt.plot(y_test, y_test, linestyle='--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Testing Data 75%)')
```
